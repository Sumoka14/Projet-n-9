```python
def checkUnusedId(dfToRmv, df2, id1, id2):
    listToRmv = dfToRmv[id1].tolist()
    listDf2 = df2[id2].tolist()
    tempo = []

    for i in listToRmv:
        if i in listDf2:
            tempo.append(i)
    for i in tempo:
        listToRmv.remove(i)

    if len(listToRmv) != 0:
        listToRmvUnique = pd.Series(listToRmv).unique()
        print(len(listToRmvUnique), id1, "non utilisé(s) (",
              len(listToRmv), ") :", listToRmvUnique)
    else:
        print("Auncun", id1, "non utilisé")
    # dfToRmv = dfToRmv.drop(dfToRmv[dfToRmv[id1].isin(listToRmv)].index)

    return listToRmv
```

Créons maintenant un tableau contenant la part de la viande animale dans la consommation en Kcal et en gramme de protéines par personne

```python
Viande_List = ['Viande de Bovins', "Viande d'Ovins/Caprins", 'Viande de porcins',
               'Viande de Suides', "Viande de Volailles", "Viande, Autre", "Abats Comestible"]
```

Calculons d'abord la consommation totale

```python
part = alim.groupby(['Pays', 'Année', 'Élément'])['Valeur'].sum().reset_index(
).pivot_table('Valeur', ['Pays', 'Année'], 'Élément').reset_index()
```

et maintenant la consommation de viande animale

```python
alimViande = alim[alim['Produit'].isin(Viande_List)]
alimViande = alimViande.groupby(['Pays', 'Année', 'Élément'])['Valeur'].sum(
).reset_index().pivot_table('Valeur', ['Pays', 'Année'], 'Élément').reset_index()
alimViande = alimViande.rename(columns={'Disponibilité alimentaire (Kcal/personne/jour)': 'Disponibilité alimentaire (Viande) (Kcal/personne/jour)',
                                        'Disponibilité de protéines en quantité (g/personne/jour)': 'Disponibilité de protéines en quantité (Viande) (g/personne/jour)'})
```

```python
part = pd.merge(part, alimViande, on=['Pays', 'Année'])
```

```python
part['Part de la viande en protéine (%)'] = round(
    100*part['Disponibilité alimentaire (Viande) (Kcal/personne/jour)']/part['Disponibilité alimentaire (Kcal/personne/jour)'], 2)
part['Part de la viande en Kcal (%)'] = round(100*part['Disponibilité de protéines en quantité (Viande) (g/personne/jour)'] /
                                              part['Disponibilité de protéines en quantité (g/personne/jour)'], 2)
```

```python
part = part[['Pays', 'Année',
             'Part de la viande en protéine (%)', 'Part de la viande en Kcal (%)']]
```

```python
part['id'] = part['Pays'] + "/" + part['Année'].astype('str')
```

```python
popPrec = pop[pop['Année'] == (annee-1)]
popActu = pop[pop['Année'] == annee]
checkUnusedId(popPrec, popActu, 'Pays', 'Pays')
checkUnusedId(popActu, popPrec, 'Pays', 'Pays')
print("Tous les pays sont bien restés")

popPrec['Population '+str(annee-1)] = popPrec.loc[:, 'Population']
popActu['Population '+str(annee)] = popActu.loc[:, 'Population']
popEvo = pd.merge(popPrec[['Pays', 'Population '+str(annee-1)]],
                  popActu[['Pays', 'Population '+str(annee)]], on='Pays')
popEvo['Population évolution (%)'] = 100*(popEvo['Population '+str(
    annee)]-popEvo['Population '+str(annee-1)])/popEvo['Population '+str(annee-1)]
```

On ignore les 17 pays qui apparaissent dans la table stabilité mais qui n'apparaissent pas dans la table part car, on n'a pas les données nécessaires sur ces pays pour mener à bien l'analyse. Pour les 4 autres (ceux qui n'apparaissent pas dans la table stabilité), ils auront la stabilité politique mediane.

```python
stabEvo
```

```python
df = pd.merge(df, stabEvo[['Pays', 'Stabilité Politique '+str(annee),
              'Stabilité Politique évolution (%)']], on='Pays', how='left')
medPrec = stabEvo['Stabilité Politique '+str(annee-1)].median()
medActu = stabEvo['Stabilité Politique '+str(annee)].median()
medPer = 100*(medActu-medPrec)/medPrec
Lmanquant = checkUnusedId(df, stabEvo, 'Pays', 'Pays')
df.loc[df['Pays'].isin(Lmanquant), 'Stabilité Politique '+str(annee)] = medActu
df.loc[df['Pays'].isin(Lmanquant),
       'Stabilité Politique évolution (%)'] = medPer
```

```python
pibChn = pd.read_csv('data\OCDE_PIB_Chine_2018_2020.csv')
pibChn = pibChn[pibChn['LOCATION'] == 'CHN']

pibChnActu = pibChn.loc[pibChn['TIME'] == annee, 'Value'].iloc[0]
pibChnPrec = pibChn.loc[pibChn['TIME'] == (annee-1), 'Value'].iloc[0]
pibChnEvo = 100*(pibChnActu-pibChnPrec)/pibChnPrec
ligne = ['Chine, continentale', pibChnPrec, pibChnActu, pibChnEvo]
pibEvo.loc[len(pibEvo)] = ligne
```
